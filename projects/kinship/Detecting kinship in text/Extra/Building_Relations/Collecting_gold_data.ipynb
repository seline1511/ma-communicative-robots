# Collecting Gold Data
This notebook uses the character relations file generated by Building_relations.ipynb to generate a list of "gold" relations (those relations that exist in the show and in the relations.pkl file represented as kinship triples).

This notebook outputs a pickle file "gold_relations.pkl" containing the list of gold relations.

```python
import pickle

class Character:
    'Class representing named characters'
    def __init__(self, firstname, lastname=None, relations=dict()):
        self.firstname = firstname
        self.lastname = lastname
        self.relations = relations
        
with open("relations.pkl", "rb") as infile:
    characters = pickle.load(infile)
```

```python
relations = []
for character in characters:
    source_name = character.firstname
    if "Mr" not in source_name:
        source_name = source_name.capitalize()
    for relation, targets in character.relations.items():
        if type(targets) != list:
            targets = [targets]
            relation_name = "has-" + relation
        else:
            relation_name = "has-" + relation[:-1]
        for target in targets:
            target_name = target.firstname
            if "Mr" not in target_name:
                target_name = target_name.capitalize()
            relations.append([source_name, relation_name, target_name])
```

```python
with open("gold_relations.pkl", "wb") as outfile:
    pickle.dump(relations, outfile)
```
